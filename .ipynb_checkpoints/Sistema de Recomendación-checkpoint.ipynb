```python
#Pandas, librería para el manejo de los dataframe dentro del programa
import pandas as pd
#Librerias de funciones matemáticas
import math
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Leer DataFrame de vinos
```

```python
vinos_df = pd.read_csv('vinosbase2.csv')
vinos_df.head()
```

```python
vinos_df['variedad'] = vinos_df.variedad.str.replace(" ","")
vinos_df['variedad'] = vinos_df.variedad.str.split(',')
vinos_df.head()
```

```python
vinos_df['aroma'] = vinos_df.aroma.str.strip()
vinos_df['aroma'] = vinos_df.aroma.str.split(',')
vinos_df.head()
```

```python
#Copying the movie dataframe into a new one since we won't need to use the genre information in our first case.
vinoVariedad_df = vinos_df.copy()

#For every row in the dataframe, iterate through the list of genres and place a 1 into the corresponding column
for index, row in vinos_df.iterrows():
    for grape in row['variedad']:
        vinoVariedad_df.at[index, grape] = 1
#Filling in the NaN values with 0 to show that a movie doesn't have that column's genre
vinoVariedad_df= vinoVariedad_df.fillna(0)
vinoVariedad_df.head()
```

```python
vinoVariedad_df.to_csv('out.csv')
```
